# Loading our trained models

```python
def load_model(model_path):
    # Load model
    model = torch.hub.load('ultralytics/yolov5', 'custom',model_path)
    model.eval()  # Set to evaluation mode
    return model
```

```python
sign_model = load_model('./best_traffic_signs.pt')
```
